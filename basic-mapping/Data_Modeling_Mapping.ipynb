## **Reflection on data models and management in ArcGIS Pro**

- **Write a 100-word paragraph which describes the relationship between**
  - a **`Geodatabase`**
  - a **`Feature Dataset`**
  - and **`Feature Classes`**
  
  Refer to **Page 9** of the assignment guide, and other resources online such as:
    - [Fundamentals of the geodatabase](https://pro.arcgis.com/en/pro-app/latest/help/data/geodatabases/overview/fundamentals-of-the-geodatabase.htm)
    - [Feature datasets in ArcGIS Pro](https://pro.arcgis.com/en/pro-app/latest/help/data/feature-datasets/feature-datasets-in-arcgis-pro.htm)

A geodatabase is a storage system of file clusters by geography, type, or other criteria. They can help organize many different types of data all together in the same database. A feature dataset is found inside a geodatabase. This is a collection of spatial data and data layers/geometric construct. Each geodatabase can house multiple feature datasets. This allows the cluster filing mentioned above to help with organization. Data layers are called feature classes, and they make up the specific cluster of data. In summary, a geodatabase is the overall container for all of the data, a feature dataset is the data sub-container found inside, and a feature class is a layer of data.
