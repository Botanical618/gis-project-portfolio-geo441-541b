## **Part 3: Reflection**

The first map shows the topography of the Oxford area. Visually, we can see the gradient for higher and lower elevation in different regions of the area. Higher elevations have a reddish color while lower areas have a bluish color. The data in the first map shows the highest and lowest values, but determining the exact elevation is not possible with this graph. It is mostly useful for seeing trends in elevation in the area. The second map shows the slope in the Oxford area. Visually, we can see each parcel in Oxford and the corresponding slope color (with a steeper slope having darker blue tones). The data in the second map gives a range of values for each shade of blue, which is useful for seeing more informative slopes in parcels. The gradient in color as well as the range of values for each shade helps visualize overall slope trends while also allowing numeric data ranges for each parcel.
